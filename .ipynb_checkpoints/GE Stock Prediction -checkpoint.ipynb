<pre><b>Making Our dataset as Supervised problem with Moving Window method</b></pre>

```python
#one month and we are considering last 30 days as features and 31 day as target
window_size = 30

# Get indices of access for the data
num_samples = len(ge) - window_size
indices = np.arange(num_samples).astype(np.int)[:,None] + np.arange(window_size + 1).astype(np.int)
```

```python
indices
```

```python
ge_mv = ge['Adj_Close'].values[indices]
X = ge_mv[:,:-1] 
y = ge_mv[:,-1]
```

```python
X.shape
```

```python
y.shape
```

<pre>It is time series data so splitting data into first 85% as training and next 15 as testing.</pre>

```python
split_indices = int(0.85 * num_samples)
X_train = X[:split_indices]
y_train = y[:split_indices]
X_test = X[split_indices:]
y_test = y[split_indices:]
```

```python
print('No of traning data samples', y_train.shape)
print('No of testing data samples', y_test.shape)
```

```python
plt.plot(y_test)
```

<pre><b>Model-1 - Lasso Regression</b></pre>

```python
from sklearn.linear_model import Ridge, Lasso, LinearRegression
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import mean_absolute_error as mae
```

```python
# Train
Lasso_model = Lasso(alpha=1.0)
Lasso_model.fit(X_train, y_train)
```

```python
y_pred_train = Lasso_model.predict(X_train)
y_pred_test = Lasso_model.predict(X_test)
```

```python
print('Mean Absolute Error for Train data', mae(y_train, y_pred_train))
print('Mean Absolute Error for Test data', mae(y_test, y_pred_test))
```

```python
lasso_df = ge.copy()
lasso_df.drop(['Open', 'High', 'Low', 'Close', 'Volume'], axis=1, inplace=True)
lasso_df = lasso_df.iloc[window_size:split_indices] # Past 32 days we don't know yet
lasso_df['Adj Close Train'] = y_pred_train[:-window_size]
lasso_df.plot(label='GE', figsize=(16,8), title='Adjusted Closing Price', grid=True)
```
